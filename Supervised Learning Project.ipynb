### E. Balance the data using the right balancing technique. [2 Marks]

### ii. Say output is class A : 20% and class B : 80%
### iii. Here you need to balance the target variable as 50:50.
### iv. Try appropriate method to achieve the same.

```python
#checking distribution of the target variable
dist=mdf['LoanOnCard'].value_counts(normalize=True)*100
print(dist)
```

```python
from sklearn.utils import resample
majority_class = mdf[mdf['LoanOnCard'] == 0]
minority_class = mdf[mdf['LoanOnCard'] == 1]

# Downsample the majority class to match the minority class
majority_downsampled = resample(majority_class, replace=False, n_samples=len(minority_class))

# Combine the minority class and downsampled majority class
balanced_df = pd.concat([majority_downsampled, minority_class])

# Check the distribution of the target variable in the balanced dataframe
target_dist = balanced_df['LoanOnCard'].value_counts(normalize=True) * 100
print(target_dist)
```

the resample() function from Scikit-learn to downsample the majority class to match the minority class. The replace=False parameter ensures that sampling is done without replacement. The n_samples parameter is set to the length of the minority class.

Finally, you combine the minority class and downsampled majority class using the pd.concat() function, and save the result in balanced_df.
